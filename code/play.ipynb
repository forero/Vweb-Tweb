```python
circ_cut = loadtxt("mu_cuts_circ_512.dat")
conc_cut = loadtxt("mu_cuts_conc_512.dat")
dens_cut = loadtxt("mu_cuts_dens_512.dat")
spin_cut = loadtxt("mu_cuts_spin_512.dat")
tri_cut = loadtxt("mu_cuts_tri_512.dat")
```

```python
def extract_across(MY_col):
    mu_high30_col = 0
    mu_low30_col = 1
    mu_all_col = 4
    AXT = zeros((5,3))
    AXT[0,0] = circ_cut[MY_col, mu_high30_col]
    AXT[0,1] = circ_cut[MY_col, mu_low30_col]
    AXT[0,2] = circ_cut[MY_col, mu_all_col] 

    AXT[1,0] = conc_cut[MY_col, mu_high30_col]
    AXT[1,1] = conc_cut[MY_col, mu_low30_col]
    AXT[1,2] = conc_cut[MY_col, mu_all_col] 

    AXT[2,0] = dens_cut[MY_col, mu_high30_col]
    AXT[2,1] = dens_cut[MY_col, mu_low30_col]
    AXT[2,2] = dens_cut[MY_col, mu_all_col] 

    AXT[3,0] = spin_cut[MY_col, mu_high30_col]
    AXT[3,1] = spin_cut[MY_col, mu_low30_col]
    AXT[3,2] = spin_cut[MY_col, mu_all_col] 

    AXT[4,0] = tri_cut[MY_col, mu_high30_col]
    AXT[4,1] = tri_cut[MY_col, mu_low30_col]
    AXT[4,2] = tri_cut[MY_col, mu_all_col]
    return AXT

```

```python
AX_T = extract_across(1)

ax = axes()
ax.set_ylim([0.44,0.56])

    
scatter(arange(5),AX_T[:,2], marker='s', s=100)
scatter(arange(5),AX_T[:,0], marker='^', s=100)
scatter(arange(5),AX_T[:,1], marker='v', s=100)
p = plt.axhspan(0.49, 0.51, facecolor='0.5', alpha=0.3)
```
