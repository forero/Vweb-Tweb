```python
n_resolutions = 2
n_web_types = 2
n_alignments = 3

resolutions = [256,512]
web_types = ["T", "V"]
alignments = ["SHAPE", "V", "J"]

filename="../data/preferential_%sweb%d_halo_%s.dat"%(web_types[0], resolutions[0], alignments[0])
raw_data = loadtxt(filename)
n_lines = size(raw_data[:,0])
n_columns = size(raw_data[0,:])

full_data = ones((n_lines, n_columns+3, n_web_types, n_resolutions, n_alignments))
MASS = 0
F_EIGEN1 = 1
F_EIGEN2 = 2
F_EIGEN3 = 3
N_TOT = 4
SIGMA_EIGEN1 = 5
SIGMA_EIGEN2 = 6
SIGMA_EIGEN3 = 7
TWEB = 0
VWEB = 1
R256 = 0
R512 = 1
A_SH = 0
A_V = 1
A_J = 2
```

```python
for n in range(n_alignments):
    for l in range(n_resolutions):
        for m in range(n_web_types):
            filename="../data/preferential_%sweb%d_halo_%s.dat"%(web_types[l], resolutions[m], alignments[n])
            raw_data = loadtxt(filename)
            n_lines = size(raw_data[:,0])
            n_columns = size(raw_data[0,:])
            for i in range(n_lines):
                for j in range(n_columns):
                    full_data[i,j,l,m,n] = raw_data[i,j]
                for j in range(n_columns, n_columns+3): #include the errors
                    full_data[i,j,l,m,n] = sqrt(raw_data[i,j-n_columns+1])
                for j in range(1,4): #normalize to get the fraction
                    full_data[i,j,l,m,n] =  full_data[i,j,l,m,n]/full_data[i,N_TOT,l,m,n]
                for j in range(5,8): #normalize to get the error
                    full_data[i,j,l,m,n] =  full_data[i,j,l,m,n]/full_data[i,N_TOT,l,m,n]
print full_data[:,MASS,0,0,0]
```

```python

WEB = VWEB
ALIGN = A_SH
def sub_plot(WEB, ALIGN):
    RES = R256
    colors = ['black', 'red', 'yellow']
    for i in range(3):
        fill_between(full_data[:,0,0,0,0], 
                 full_data[:,F_EIGEN1+i,WEB,RES,ALIGN] - full_data[:,SIGMA_EIGEN1+i,WEB,RES,ALIGN], 
                 full_data[:,F_EIGEN1+i,WEB,RES,ALIGN] + full_data[:,SIGMA_EIGEN1+i,WEB,RES,ALIGN],
                 alpha=0.8, facecolor=colors[i],label='as')
        fill_between(full_data[:,0,0,0,0], 
                 full_data[:,F_EIGEN1+i,WEB,RES+1,ALIGN] - full_data[:,SIGMA_EIGEN1+i,WEB,RES+1,ALIGN], 
                 full_data[:,F_EIGEN1+i,WEB,RES+1,ALIGN] + full_data[:,SIGMA_EIGEN1+i,WEB,RES+1,ALIGN],
                 alpha=0.4, facecolor=colors[i],label='as')

    line_y = ones(n_lines) /3.0
    plot(full_data[:,0,0,0,0], line_y, color='black', linewidth=5)
```

```python
def enlarge_labels(ax):
    ticklabels_x = ax.get_xticklabels()
    ticklabels_y = ax.get_yticklabels()
    for label_x in ticklabels_x:
        label_x.set_fontsize(20)
        label_x.set_family('serif')
    for label_y in ticklabels_y:
        label_y.set_fontsize(20)
        label_y.set_family('serif')
```
