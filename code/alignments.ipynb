```python
n_points = 300
phi = 2.0*pi*random.random(n_points)
theta = arccos((random.random(n_points)-0.5)*2.0)
rand_x = sin(theta)*cos(phi)
rand_y = sin(theta)*sin(phi)
rand_z = cos(theta)
```

```python
cos_alpha = rand_x[:]
cos_beta = rand_y[:]
cos_gamma = rand_z[:]
```

```python
plot(cos_alpha**2)
```

```python
id_alpha = where((cos_alpha**2 > cos_beta**2) & (cos_alpha**2 > cos_gamma**2))
plot(cos_alpha[id_alpha]**2)
```

```python
id_beta = where((cos_beta**2 > cos_alpha**2) & (cos_beta**2 > cos_gamma**2))
plot(cos_beta[id_beta]**2)
```

```python
id_gamma = where((cos_gamma**2 > cos_alpha**2) & (cos_gamma**2 > cos_beta**2))
plot(cos_gamma[id_gamma]**2)
```

```python
scatter(rand_x[id_alpha], rand_y[id_alpha])
```
